```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#BAsic Array plot
x=np.arange(0,10)
y=2*x

print(x)
print(y)
```

```python
#Line Plot
plt.plot(x,y)
plt.xlabel('X Axis title here')
plt.ylabel('y Axis title here')
plt.title('Plot title here')
plt.show() #Required for non-jupyter users,but also removes Out[] info
```

```python
#More features 
plt.plot(x,y)
plt.xlabel('X Axis title here')
plt.ylabel('y Axis title here')
plt.title('Plot title here')
plt.xlim(0,6) #Lower limit,upper limit
plt.ylim(0,12) #Lower limit,upper limit
plt.show()
```

```python
#Exporting a plot
plt.plot(x,y)
plt.savefig('example.png')
```

```python
#Creating figure 
#the main idea in using the more formal oob method is to create a figure object and then just call methods or attrributes off of that object
#This approach is nicer when dealing with a canvas that has multiple plots on it.
```

```python
#Create an empty canvas
fig=plt.figure()
#Add set of axes of figure
axes = fig.add_axes([0,0,1,1]) #left,bottom,width,height (range 0 to 1)
#plot on that set of axes
axes.plot(x,y)
plt.show()
```

```python
fig = plt.figure()
axes=fig.add_axes([0,0,1,1]) #left,bottom,width,height (range 0 to 1)
axes.plot(a,b)
plt.show()
```

```python
#Adding another set of axes to the figure
#create blank figure
fig=plt.figure()

axes1 = fig.add_axes([0,0,1,1]) #Large figure
axes2 = fig.add_axes([0.2,0.2,0.5,0.5]) #smalller figure


axes1.plot(a,b)
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Big figure')


axes2.plot(a,b)
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes2.set_title('small figure');
plt.show()
```

```python
#Let's move the small figure and edit its parameters
fig=plt.figure()
axes1 = fig.add_axes([0,0,1,1]) #Large figure
axes2 = fig.add_axes([0.2,0.2,0.25,0.25]) #smaller figure

axes1.plot(a,b)
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')

axes1.set_title('Big figure')


axes2.plot(a,b)
axes2.set_xlim(8,10)
axes2.set_ylim(4000,10000)
axes2.set_xlabel('X Label')
axes2.set_ylabel('Y Label')
axes2.set_title('Zoomed In');
plt.show()
```

```python
a=np.linspace(0,10,11)
b=a**4
print(a)
print(b)

x=np.arange(0,10)
y=2*x
print(x)
print(y)
```

```python
# The plt.subplots object will act as a more automatic axis manager
# This makes it much easier to show multiple plots side by side
fig,axes=plt.subplots()

#Now use the axes object to add stuff to plot
axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title'); #; hides Out[]
plt.show()

```

```python
# Empty canvas of 1 by 2 subplots
fig,axes=plt.subplots(nrows=1,ncols=2)
print(axes)
print(axes.shape)
plt.show()
```

```python
# Empty canvas of 2 by 2 subplots
fig,axes=plt.subplots(nrows=2,ncols=2)
print(axes)
print(axes.shape)
plt.show()
```

```python
# Use .subplots_adjust to adjust spacing manually
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,8))

# Parameter at the axes level
axes[0][0].plot(a, b)
axes[1][1].plot(x, y)
axes[0][1].plot(y, x)
axes[1][0].plot(b, a)

# Use left, right, top, bottom to stretch subplots
# Use wspace, hspace to add spacing between subplots
fig.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.9, hspace=0.1)
plt.show()
```

```python
x=np.arange(0,10)
y=2*x

#Legends
# We can use the **label="label text"** keyword argument whn plots or other objects are added

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label="x**2")

ax.plot(x,x**3,label="x**3")

ax.legend()

plt.show()
```

```python
#fig.subplots_adjust(left=None,
fig,ax = plt.subplots()
ax.plot(x,x**2,'b,-')
ax.plot(x,x**3,'g--')
plt.show()
```

```python
fig,ax=plt.subplots()
ax.plot(x,x+1,color="blue",alpha=0.5)
ax.plot(x,x+2,color="#8B008B")
ax.plot(x,x+3,color="#FF8C00")
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(12,6))
ax.plot(x,x-1,color="blue",lw=0.5)
ax.plot(x,x-2,color="blue",lw=1)
ax.plot(x,x-3,color="blue",lw=10)
ax.plot(x,x-4,color="blue",lw=20)
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(12,6))
ax.plot(x,x-1,color="blue",lw=3,linestyle='-')
ax.plot(x,x-2,color="blue",lw=3,linestyle='-.')
ax.plot(x,x-3,color="blue",lw=3,linestyle=':')
ax.plot(x,x-4,color="blue",lw=3,linestyle='--')
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(12,6))
ax.plot(x,x-1,color="blue",marker="*",markersize=20)
ax.plot(x,x-2,color="pink",marker="o",markersize=20)
ax.plot(x,x-3,color="black",marker="s",markersize=20,lw=0)
ax.plot(x,x-4,color="red",marker="1",markersize=20)
plt.show()
```
