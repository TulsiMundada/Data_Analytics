```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_filtered = pd.DataFrame({'opposition' : ['India' , 'Australia' , 'England','South Africa'],
                            'runs_scored' : [250 , 300 , 220 , 200]})
df_grouped = df_filtered.groupby('opposition')['runs_scored'].sum()
plt.figure(figsize=(12,6))
plt.pie(df_grouped , labels=df_grouped.index , autopct ='%1.1f%%')
plt.title('total runs scored against oppositon')
plt.show()                
```
