```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data = df)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data = df,hue='gender')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='parental level of education',y='math score',data = df,hue='gender',split=True ,palette='viridis')
plt.show()
```

```python
sns.swarmplot(x='math score',data=df)
plt.show()
```

```python
sns.swarmplot(x='math score',data=df,size=2)
plt.show()
```

```python
sns.swarmplot(x='math score',y='race/ethnicity',data=df,size=3)
plt.show()
```

```python
sns.swarmplot(x='race/ethnicity',y='math score',data=df)
plt.show()
```

```python
sns.boxenplot(x='math score',y='race/ethnicity',data=df)
plt.show()
```

```python
sns.boxenplot(y='math score',x='race/ethnicity',data=df,palette='viridis')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxenplot(y='math score',x='race/ethnicity',data=df,hue='gender',palette='viridis')
plt.show()
```
