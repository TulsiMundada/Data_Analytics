### 2.Create Winning seats distribution by Major Political Parties & Alliances for 2009 & 2014

```python
df1_winners = df1[df1['Position'] == 1]
df3_winners = df3[df3['Position'] == 1]
major_parties = ['BJP,INC','BSP']
# Now we can group the winning candidates by their party and grand alliance and count the number of seats won.
df1_seats_by_party = df1_winners.groupby(['Party Abbreviation', 'Grand Alliance'])['Position'].count()
df3_seats_by_party = df3_winners.groupby(['Party Abbreviation', 'Grand Alliance'])['Position'].count()

#display the results.
print("Winning seats distribution by Major Political Parties & Alliances for 2009:")
print(df1_seats_by_party)

print("\nWinning seats distribution by Major Political Parties & Alliances for 2014:")
print(df3_seats_by_party)
```

```python
import matplotlib.pyplot as plt

labels = df1_seats_by_party.index.to_list()
values = df1_seats_by_party.values.tolist()
print(labels)
print(values)
# Create bar chart
plt.bar(range(len(values)), values, tick_label=labels)
plt.xlabel("Grand Alliance")
plt.ylabel("Number of Seats Won")
plt.title("Winning Seats Distribution by Grand Alliance in 2009")
plt.show()
```

```python
#4.Plot comparatively seats won based on candidate category as General, ST and SC for 2009 & 2014

import matplotlib.pyplot as plt
#Grouping all the category seatd together
df1_seats_by_category = df1_winners.groupby(['Candidate Category'])['Position'].count()
df3_seats_by_category = df3_winners.groupby(['Candidate Category'])['Position'].count()

#display the results.
print("Winning seats distribution by category for 2009:")
print(df1_seats_by_category)

print("\nWinning seats distribution by category for 2014:")
print(df3_seats_by_category)

# Create a bar chart of the number of seats won by each party in 2009
df1_seats_by_category.plot(kind='bar')
plt.title('Winning seats distribution by category for 2009')
plt.xlabel('Category')
plt.ylabel('Number of seats')
plt.show()

# Create a bar chart of the number of seats won by each party in 2014
df3_seats_by_category.plot(kind='bar')
plt.title('Winning seats distribution by category for 2014')
plt.xlabel('Category')
plt.ylabel('Number of seats')
plt.show()
```

```python
Plot the age distribution of winners of both 2014 and 2009 elections

import matplotlib.pyplot as plt
# Create a histogram of the age distribution of winners in 2009
df1_winners['Candidate Age'].plot.hist(bins=10)
plt.title('Candidate Age distribution of winners in 2009')
plt.xlabel('Age')
plt.ylabel('Number of winners')
plt.show()

# Create a histogram of the age distribution of winners in 2014
df3_winners['Candidate Age'].plot.hist(bins=10)
plt.title('Candidate Age distribution of winners in 2014')
plt.xlabel('Age')
plt.ylabel('Number of winners')
plt.show()
```
